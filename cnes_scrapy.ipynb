### Tabela Dialise

```python
import pandas as pd

# Inicialize o DataFrame geral fora da função
df_geral = pd.DataFrame(columns=['CNES', 'Salas Hemodiálise', 'Salas de Reuso', 'Máquinas Hemodiálise', "Tratamento D'água"])

def tab_caracterizacao_servico_dialise(cnes):
    # Localize a tabela usando um seletor CSS ou XPath
    tabela = driver.find_element(By.XPATH, '//table[@class="table table-bordered table-striped table-condensed"]')

    # Encontre todas as linhas da tabela no corpo do tbody
    linhas = tabela.find_elements(By.XPATH, './/tbody/tr')

    # Itere pelas linhas da tabela
    for linha in linhas:
        # Encontre as células (colunas) dentro de cada linha
        colunas = linha.find_elements(By.XPATH, './/td')
        
        # Crie um DataFrame temporário com os dados da tabela atual
        caracterizacao_servico_dialise = pd.DataFrame({
            'CNES': [cnes],
            'Salas Hemodiálise': [colunas[0].text],
            'Salas de Reuso': [colunas[1].text],
            'Máquinas Hemodiálise': [colunas[2].text],
            "Tratamento D'água": [colunas[3].text]
        })

        global df_geral  # Acesso ao DataFrame global
        df_geral = pd.concat([df_geral, caracterizacao_servico_dialise], ignore_index=True)
    display(df_geral)
```
