#### Caracterização de serviço de diálise

```python
# Inicialize o DataFrame geral fora da função
df_caracterizacao_servico_dialise = pd.DataFrame(columns=['CNES', 'Salas Hemodiálise', 'Salas de Reuso', 'Máquinas Hemodiálise', "Tratamento D'água"])

def tab_caracterizacao_servico_dialise(cnes):

    # Localize a tabela usando um seletor CSS ou XPath
    tabela = driver.find_element(By.XPATH, '//table[@class="table table-bordered table-striped table-condensed"]')

    # Encontre todas as linhas da tabela no corpo do tbody
    linhas = tabela.find_elements(By.XPATH, './/tbody/tr')
    
    # Itere pelas linhas da tabela
    for linha in linhas:
        # Encontre as células (colunas) dentro de cada linha
        colunas = linha.find_elements(By.XPATH, './/td')
        
        # Handle the checkbox status
        checkbox = linha.find_element(By.XPATH, './/input[@type="checkbox"]')
        flag = checkbox.is_selected()  # Returns True if checked, False otherwise

        # Crie um DataFrame temporário com os dados da tabela atual
        caracterizacao_servico_dialise = pd.DataFrame({
            'CNES': [cnes],
            'Salas Hemodiálise': [colunas[0].text],
            'Salas de Reuso': [colunas[1].text],
            'Máquinas Hemodiálise': [colunas[2].text],
            "Tratamento D'água": [colunas[3].text],
            "Flag": [flag]
        })

        global df_caracterizacao_servico_dialise  # Acesso ao DataFrame global
        df_caracterizacao_servico_dialise = pd.concat([df_caracterizacao_servico_dialise, caracterizacao_servico_dialise], ignore_index=True)

    #print('Caracterização de serviço de diálise')
    #display(df_caracterizacao_servico_dialise)
```

#### Serviço de referência e manutenção

```python
df_servico_referencia_manutencao = pd.DataFrame(columns=['CNES', "Servico", "Razao Social", "CNPJ", "Municipio"])

def tab_servico_referencia_manutencao(cnes):
    global df_servico_referencia_manutencao
    lista = []
    for i in range(1,6):
        try:
            row_xpath = f'//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[2]/table/tbody/tr[{i}]'
            Servico = driver.find_element(By.XPATH, f'{row_xpath}/td[1]').text
            Razao_Social = driver.find_element(By.XPATH, f'{row_xpath}/td[2]').text
            CNPJ = driver.find_element(By.XPATH, f'{row_xpath}/td[3]').text
            Municipio = driver.find_element(By.XPATH, f'{row_xpath}/td[4]').text

            lista.append({'CNES': cnes, "Servico": Servico, "Razao Social": Razao_Social, "CNPJ": CNPJ, "Municipio": Municipio})
        except:
            break
          
    temp_df = pd.DataFrame(lista)

    # Concatena com o DataFrame global
    df_servico_referencia_manutencao = pd.concat([df_servico_referencia_manutencao, temp_df], ignore_index=True)
      
      
def func_loop_servico(cnes):
    global df_servico_referencia_manutencao
    for i in range(3, 12):
        try:
            tab_servico_referencia_manutencao(cnes)
            # time.sleep(3.5)
            # Inserido num_page_topic pois ele meio que 'acalma' o selenium, para nao repetir a coleta dos dados
            num_page_topic = driver.find_element(By.XPATH, f'//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[2]/div/div/div/ul/li[{i}]/a/span').text
            button_xpath = f'//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[2]/div/div/div/ul/li[{i}]/a'
            next_button = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, button_xpath)))
            next_button.click()
        except NoSuchElementException:
            break
        except TimeoutException:
            break                          

    #print('Serviço de referência e manutenção')
    #display(df_servico_referencia_manutencao)
```

```python
def get_values_dialise(cnes):

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[1]/table/thead/tr[1]/th')))
    caracterizacao_servico_dialise = driver.find_element(By.XPATH, '//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[1]/table/thead/tr[1]/th').text
    servico_referencia = driver.find_element(By.XPATH, '//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[2]/table/thead/tr[1]/th').text
    formalizacao = driver.find_element(By.XPATH, '//*[@id="estabContent"]/div/section/div[3]/div/div[2]/div[1]/div[3]/table/thead/tr/th').text
    
    lista = [caracterizacao_servico_dialise, servico_referencia, formalizacao]

    for i in lista:
        if i == 'Caracterização de serviço de diálise':
            tab_caracterizacao_servico_dialise(cnes)
        elif i == 'Serviço de referência e manutenção':
            func_loop_servico(cnes)
        elif i == 'Formalização':
            tab_formalizacao(cnes)
```
